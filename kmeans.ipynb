```python
df[df["Agriculture"].isna()] = df[df["Agriculture"].isna()].fillna(0)
```

```python
df[df["Literacy (%)"].isnull()]
```

```python
df.groupby("Region")["Literacy (%)"].mean().sort_values(ascending=False)
```

```python
#ülkelerde olmayan verileri bir üsteki satırdaki bilgilere göre doldurucam

df["Literacy (%)"] = df["Literacy (%)"].fillna(df.groupby("Region")["Literacy (%)"].transform("mean"))
```

```python
df.isna().sum()
```

```python
df["Climate"] = df["Climate"].fillna(df.groupby("Region")["Climate"].transform("mean"))
```

```python
df.isna().sum()
```

```python
# Tüm sayısal sütunlarda eksik değerleri doldurma
numeric_cols = df.select_dtypes(include=['number']).columns
df[numeric_cols] = df[numeric_cols].fillna(df.groupby('Region')[numeric_cols].transform('mean'))
```

```python
X= df.drop("Country", axis=1)
```

```python
X= pd.get_dummies(X, drop_first=True)
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaled_X = StandardScaler().fit_transform(X)
```

```python
from sklearn.cluster import KMeans
```

```python
ssd = []

for k in range(2,50):
    model = KMeans(n_clusters=k)
    model.fit(scaled_X)
    ssd.append(model.inertia_)
```

```python
plt.plot(range(2,50),ssd, "o" )
plt.xlabel("K")
plt.ylabel("noktalar arası uzaklığın karelerinin toplamı")
```

```python
model = KMeans(n_clusters=3)
model.fit(scaled_X)
```

```python
model.labels_
```

```python
X["Cluster"] = model.labels_
```

```python
plt.figure(figsize=(12,9),dpi=200)
sns.barplot(x=X.columns, y=X.corr()["Cluster"].sort_values())
plt.xticks(rotation=90);
```
